```python
# first, make a yaml file with all the species in the database on it

entries = []
for i in range(len(database_fun.total_species_list)):
    sp = database_fun.index2species(i)
    entry = {}
    entry['name'] = str(sp)
    entry['type'] = 'species'
    entry['index'] = i
    entry['complete'] = False
    entries.append(entry)
with open('/work/westgroup/harris.se/autoscience/reaction_calculator/dft/all_species.yaml', 'w') as f:
    yaml.dump(entries, f)
```

```python
# copy the reactions over

# next, we'll try copying over each completed species
print(f'old index\tnew index')
for old_reaction_index in range(0, 2500):
    # the indices for the species are all the same, but we'll write this code anyways for easy copying when it's time to do the reactions
    old_reaction_dir = os.path.join(database_fun.DFT_DIR, 'kinetics', 'kinetics_20230816', f'reaction_{old_reaction_index:04}')
    if not os.path.exists(old_reaction_dir):
        continue
    
    old_smiles = thermokinetic_fun.reaction_index2smiles(old_reaction_index)
    old_reaction = thermokinetic_fun.smiles2reaction(old_smiles)
    
    new_reaction_index = database_fun.get_unique_reaction_index(old_reaction)
    new_reaction_dir = os.path.join(database_fun.DFT_DIR, 'kinetics', f'reaction_{new_reaction_index:06}')
    print(old_reaction_index, '\t', new_reaction_index)
    shutil.copytree(old_reaction_dir, new_reaction_dir)
    
#     break
```
